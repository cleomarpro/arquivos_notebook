```python
#1 - acessar um site
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
headers = {'User-Agent':"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36"}
url = "https://www.google.com/search?q=tabela+do+brasileirao&oq=tabela+do+&aqs=edge.1.69i57j0i3j0i433i512j0i512l4j0i131i433i512j0i131i433.6711j0j4&sourceid=chrome&ie=UTF-8#sie=lg;/g/11jspy1hvm;2;/m/0fnk7q;st;fp;1;;;"
dados_da_pag = requests.get(url=url, headers = headers)
dados_da_pag = BeautifulSoup(dados_da_pag.content, 'html.parser')
tabela = dados_da_pag.find(class_="Jzru1c")
coluna_da_tabela = tabela.find_all('span', class_="YCyuEf")
detalhes_dos_jogos = tabela.find_all('tr', jscontroller="cvgK0e")
times = tabela.find_all('span', class_="ellipsisize")

detalhes_dos_jogos = [['1América-MG00000000'], ['2Athletico-PR00000000'], ['3Atlético-MG00000000'], ['4Bahia00000000']]

coluna_tabela = []
for coluna in coluna_da_tabela:
    coluna_tabela += coluna.contents
#print( coluna_tabela)

detalhe_do_jogo = []
for n in range(4):
    detalhes_jogos = str(detalhes_dos_jogos[n]).replace('[','').replace(']','').replace('-','').replace("'","")
    detalhes_jogos = list(detalhes_jogos)
    detalhe_do_jogo.append(detalhes_jogos[-8:])
#print(detalhe_do_jogo)

times=[]
for n in range(4):
    times_list = str(detalhes_dos_jogos[n]).replace('[','').replace(']','').replace("'","")
    times += [times_list[1:-8]]
#print(times)

detalhe_do_jogos = []
for n in range(4):
    detalhe_do_jogos = detalhe_do_jogo
    detalhe_do_jogos[n].insert(0, times[n])

coluna_tabela.insert(0, 'Times')
pd.DataFrame(detalhe_do_jogo, columns = coluna_tabela)
```
