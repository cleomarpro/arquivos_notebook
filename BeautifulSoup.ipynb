```python
#1 - acessar um site
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
headers = {'User-Agent':"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36"}
url = "https://www.google.com/search?q=tabela+do+brasileirao&oq=tabela+do+&aqs=edge.1.69i57j0i3j0i433i512j0i512l4j0i131i433i512j0i131i433.6711j0j4&sourceid=chrome&ie=UTF-8#sie=lg;/g/11jspy1hvm;2;/m/0fnk7q;st;fp;1;;;"
dados_da_pag = requests.get(url=url, headers = headers)
dados_da_pag = BeautifulSoup(dados_da_pag.content, 'html.parser')
tabela = dados_da_pag.find(class_="Jzru1c")# Pega uma teg de id, classe especifica <span class="ellipsisize hsKSJe">América-MG</span>
coluna = tabela.find_all('span', class_="YCyuEf")
coluna2 = tabela.find_all('tr', jscontroller="cvgK0e")
linha = tabela.find_all('span', class_="ellipsisize")


li= []
for linha in linha:
    li += linha.contents
print(li)
print('/////////')
col = []
for coluna in coluna:
    col += coluna.contents
    #col += coluna.split()
print( col)

col2 = []
for coluna in coluna2:
    col2 += coluna
print( col2)
```
