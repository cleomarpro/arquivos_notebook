```python
import pandas as pd

#passo 1:  Importar a base de dados
tabela = pd.read_csv(r"arquivos\telecom_users.csv")
#passo 2: visualizar a base de dados
#display(tabela)
# - intender as informações que está disponível
# - descobrir as erros da base de dados e tratar 
tabela = tabela.drop("Unnamed: 0", axis=1) # excluir coluna
tabela["TotalGasto"]=pd.to_numeric(tabela["TotalGasto"], errors="coerce") # transforma os dados da coluna em floate, em errors podecolocar o ignore->ingorar os erros, raise-> mostrar erro, coerce -> forcar a barra
tabela = tabela.dropna(how='all', axis=1) # excluir counas que esteja totalmete vazias, how = 'all' -> colunas vazias
tabela = tabela.dropna(how='any', axis=0) # excluir linhas que tiver algum campo vazio, how='any' -> algum campo vazio
display(tabela)
```

```python



print(tabela.info())
```

```python
#passo 4: análise inicial
print(tabela["Churn"].value_counts())
print(tabela["Churn"].value_counts(normalize=True).map("{:.2%}".format))
```

```python
# criar gráfico
import plotly.express as px
for coluna in tabela.columns:
    grafico = px.histogram(tabela, x=coluna, color= "Churn", text_auto=True)
    #exibir grafico
    grafico.show()
```
